```python
#### Removing Links

df['new_description'] = df['description'].str.replace(r"http\S+", "") 
df.head()
```

```python
#### removing punctuations,numbers,special character
df['new_description'] = df['description'].str.replace("[^a-zA-Z]", " ") 
```

```python
df.head()
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    clean_tweet =' '.join([word for word in text.split() if word not in stopwords])
    return clean_tweet
```

```python
df['new_description'] = df['new_description'].apply(lambda text : remove_stopwords(text.lower()))
df.head()
```

```python
 nltk.download('punkt')
```

```python
#Tokenization
df['new_description'] = df['new_description'].apply(lambda x: x.split())
df.head()
```

```python
#Stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer() 
```

```python
df['new_description'] = df['new_description'].apply(lambda x: [stemmer.stem(i) for i in x])
df.head()
```

```python
#Lemmantization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def word_lemmatizer(text):
    lem_text = [lemmatizer.lemmatize(i) for i in text]
    return lem_text
```
